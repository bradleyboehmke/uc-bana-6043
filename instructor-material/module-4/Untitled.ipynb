# Knowledge check

```python
import pandas as pd
from completejourney_py import get_data

cj_data = get_data()
products = cj_data['products']
```

1. How many products contain the word “bulk” in `product_type`?

```python
products[products['product_type'].str.contains("bulk", case=False, na=False)]
```

2. How many products do not contain punctuation in their `package_size`?

```python
contain_puncts = products['package_size'].str.contains("[.,(?>]", case=False, na=False)
products[contain_puncts]
```

3. Find all frozen pizza products.

```python
pizza = products['product_category'].str.contains("pizza", case=False, na=False)
frozen = products['product_category'].str.contains("frozen", case=False, na=False)
products[pizza & frozen]['product_category'].value_counts()
```
