## Knowledge check

Re-write the following for loop using a dictionary comprehension:

````python
import glob
import os
import pandas as pd

monthly_data_files = glob.glob("monthly_data/*")
````

````python
monthly_data = {}

for file in monthly_data_files:
    file_name = os.path.basename(file)[:-4]
    monthly_data[file_name] = file
    
monthly_data
````

````python
monthly_data = {os.path.basename(file)[:-4]: pd.read_csv(file) for file in monthly_data_files}
monthly_data
````

````python
pd.concat(monthly_data)
````
